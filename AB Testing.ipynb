Conjugate Prior of Bernoulli Likelihood
--------
This section demonstrates how posterior distribution of click through rate (CTR) adapt as we continue experienment

Import required packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import beta
```

Assuming the true click through rate is p, let start up by assign a value to p.
Define a function to run a single experiment and return the result

```python
def newexp(p):
    return np.random.random()<p
```

```python
##test newexp()
n_sample = 10000
p_test=0.9
testseq=pd.Series(newexp(p_test) for i in range(n_sample))
sample_rate=sum(testseq)/len(testseq)
print(sample_rate)  ##sampe rate should be reasonabaly close to p for large sample size
```

The prior distribution of p before we start the experienment is a Uniform distribution unif(0,1), assuming we don't have much info regarding the CRT at the beginning. Note that unif(0,1) happens to be the same as Beta Distribution, Beta(1,1). From Bayesian Statistics, we know the the posterier is also a beta distribution. And it can be shown that the updated Beta parameters, i.e. alpha and beta, can be easily calculated. New alpha = Old alpha + #of Success in the new set of experiment. New beta = old beta + #of falure in the new set of experienment. 

Let's establish some intuitive behind the parameter update for Beta distribution, without digging into the math. For a beta distribution, parameter alpha and beta together determine the shape of the pdf. The mean of the distribution is determined by the ratio alpha/beta, a success event, will push this ratio up(alpha increase by 1 while beta increase by 0) and result in a high mean. In our case of CTR, the mean simply is the current view on expected CTR. This make sense because a success event does give some incentive to think the actual CTR will a bit higher. Now how about the variance? the variance of a Beta distribution, will desease when alpha or beta increases. So either a success event, in which we increase alpha, OR a failure event, in which we increase beta, will result in a low variance. this also make sense intuitively because as we continue doing more experiment, we are more certain about the the CRT rate, i.e. the reduce in diversion of the distribution.

Define a function to update the beta parameter

```python
def updatebeta(a, b, n_suc):
   a = a + n_suc 
   b = b + (1 - n_suc)
   return a, b
```

```python
##test updatebeta()
x,y=1,1
p_test=0.5

print('x is '+str(x)+' y is '+str(y));
n_suc_test = newexp(p_test)
x,y= updatebeta(x,y,n_suc_test)
print('Outcome is '+str(n_suc_test))
print('x is '+str(x)+' y is '+str(y));
```

Now let us run some experienment and see what the posterier distribution looks like

```python
%matplotlib inline
a=1; b=1 ##initialize alpha, beta
p=0.4 ##True CTR
n=10000
nlist=np.array([10,50,100,500,1000])
x=np.linspace(0,1,200) 
for i in range(n):
    n_suc = newexp(p)
    a,b= updatebeta(a,b,n_suc)
    if i in nlist:
        y=beta.pdf(x,a,b)
        plt.plot(x,y,label="Distribution after %s trials" %i)
        plt.title("The posterier distribution of CTR")
        plt.ylim(0, 30)
        plt.legend()
plt.show()
```

As expected, the posterier distribution grows sharper as the number of experienment increases. The mean of the beta distribution are converging to the real click through rate. This suggest that the posterier distribution provide a great model of CTR as #of experiments grows.

To put it together, below is a cleaner version of the code with the help of python class. 

```python
class coingame:
    
    def __init__(self,p):
        self.p = p
        self.a = 1
        self.b = 1
        self.r = 0 ##default initial result
        self.rlist = np.empty(0) ##array to record result history
        self.alist = np.empty(0)
        self.blist = np.empty(0)
        
    def toss(self):
        self.r = (np.random.random()<self.p)
        self.rlist = np.append(self.rlist,self.r)
        return self.r
    
    def viewupdate(self,x):
        self.a = self.a + x
        self.b = self.b + 1 - x
        self.alist = np.append(self.alist,self.a)
        self.blist = np.append(self.blist,self.b)
```

```python
%matplotlib inline
n = 1000
p = 0.7
coinx = coingame(p)
x = np.linspace(0,1,200)

for i in range(n):
    coinx.toss()
    coinx.viewupdate(coinx.r)
    y = beta.pdf(x,coinx.a,coinx.b)
    
plt.plot(x,y,label='Posterier PDF',color='k')
plt.title("The posterier distribution of CTR after %s trials" %n)
plt.axvline(x=coinx.p,color='r')
plt.show()
plt.plot(coinx.alist/coinx.blist,color='g')
plt.title("Value of a/b as a function of number of trials")
plt.show()
```

Note how the value of a/b changes as we step through the experienment. The value show some significant fluctuation as the start and appears to converging to some constant. Remember I metioned earlier a/b determine the mean of beta distribution, thus the mean of the distribution will also show a similar pattern, the mean is converging to actual CTR that we used to simulate the data.
